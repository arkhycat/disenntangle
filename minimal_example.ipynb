```python
import numpy as np
import torch

a = torch.arange(6., 12.)
torch.reshape(a, (3, 2))
```

```python
from torch.utils.data import Dataset, DataLoader

class MinimalDataset(Dataset):
    def __init__(self):
        a = torch.arange(4., dtype=torch.float)
        self.a = torch.reshape(a, (2, 2))
        b = torch.arange(6., 12.)
        self.b = torch.reshape(b, (3, 2))

    def __len__(self):
        return 10000
    
    def __getitem__(self, idx):
        input = torch.randint(0, 100, size=(2,), dtype=torch.float)
        x = self.a.matmul(input)
        y = self.b.matmul(input) < 900
        y_bool = torch.logical_or(torch.logical_and(y[0], y[1]), torch.logical_and(torch.logical_not(y[0]), y[2]))
        return x, self.b.matmul(input)
```

```python
trainloader = torch.utils.data.DataLoader(MinimalDataset(), batch_size=32)
for data in trainloader:
    break
```

```python
import torch.nn as nn

class Net(nn.Module):
    def __init__(self):
        super().__init__()
        self.l1 = nn.Linear(2, 100, bias = True)
        self.l2 = nn.Linear(100, 100, bias = True)
        self.l3 = nn.Linear(100, 100, bias = True)
        self.l4 = nn.Linear(100, 100, bias = True)
        self.l5 = nn.Linear(100, 100, bias = True)
        self.l6 = nn.Linear(100, 3, bias = True)
    
    def forward(self, x):
        m = nn.ReLU()
        x = self.l1(m(x))
        x = self.l2(m(x))
        x = self.l3(m(x))
        x = self.l4(m(x))
        x = self.l5(m(x))
        x = self.l6(m(x))
        return x
```

```python

import torch.optim as optim
criterion = nn.CrossEntropyLoss()
model = Net()
optimizer = optim.SGD(model.parameters(), lr=0.1, momentum=0.9)
train_running_loss = 0
train_running_correct = 0

for i, (data, target) in enumerate(trainloader):
    optimizer.zero_grad()
    output = model(data)
    loss = criterion(output, target)# + block_reg
    train_running_loss += loss.item()
    _, preds = torch.max(output.data, 1)
    train_running_correct += (preds == target).sum().item()
    loss.backward()
    optimizer.step()
    #print(loss)
    #print(preds, target)

train_loss = train_running_loss/len(trainloader.dataset)
train_accuracy = 100. * train_running_correct/len(trainloader.dataset)
print(f'Train Loss: {train_loss:.4f}, Train Acc: {train_accuracy:.2f}')
```
